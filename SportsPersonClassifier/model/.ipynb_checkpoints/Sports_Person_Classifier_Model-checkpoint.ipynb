<h3 style='color:green'> Detect the face/eyes using HaarCascade API</h3>

```python
#use haarcascade api(downloaded from Github) to detect the face and save face details in the variable#
face_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')
faces = face_cascade.detectMultiScale(gray,1.3,5)
faces
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
#mark rectangle over the face area in the image
face_img  = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)

#plot the image
plt.imshow(face_img)
```

```python
#Detecting face/eyes with rectangular - Using HaarCascade
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),5)

plt.figure()
plt.imshow(face_img,cmap ='gray')
plt.show()
```

<h3 style='color:green'> (2) Preprocessing: Crop the facial region of the image </h3>
Load image, detect face. If eyes >=2, then save and crop the face region

```python
# write a function to crop the face from mutliple images

def get_cropped_images_if_2_eyes(img_path):
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        for (x,y,w,h) in faces:
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            if len(eyes)>=2:
                return roi_color
```

```python
#assigning the source and target folders in the 2 different variables
path_to_data = "./dataset/"
path_to_cr_data = "./dataset/cropped/"
```

```python
#use lib "os" to initialise all directories in the array
import os
img_dirs =[]
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
img_dirs
```

```python
import shutil
if os.path.exists(path_to_cr_data):
    shutil.rmtree(path_to_cr_data)
os.mkdir(path_to_cr_data)
```

```python
#write function to crop face from the image dataset
cropped_images_dir = []
celebrity_file_names_dict = {}

for img_dirs in img_dirs:
    celebirity_name = img_dirs.split('/')[-1]
    print(celebirity_name)
    
    celebrity_file_names_dict[celebirity_name]=[]
    count = 1
    
    for entry in os.scandir(img_dirs):
        roi_color = get_cropped_images_if_2_eyes(entry.path)
        if roi_color is not None:
            cropped_folder = path_to_cr_data + celebirity_name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_images_dir.append(cropped_folder)
                print("Folder Generated", cropped_folder)
                
            cropped_file_name = celebirity_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name
            
            cv2.imwrite(cropped_file_path, roi_color)
            celebrity_file_names_dict[celebirity_name].append(cropped_file_path)
            count+=1
```

<h3 style='color=green'> (3) Preprocessing: Use wavelet transform as a feature for traning our model </h3>
Wavelet transform


```python
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
img_haar = w2d(cropped_image,'db1',5)
plt.imshow(img_haar, cmap='gray')
```

```python
class_dict = {}
count = 0
for celebirity_name in celebrity_file_names_dict.keys():
    print(celebirity_name)
    class_dict[celebirity_name] = count
    count = 1+count
class_dict
```

<h3 style ='color:green'> creating x and y values for tarining model</h3>

```python
X =[]
y =[]

for celebirity_name, training_files in celebrity_file_names_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
        if img is None:
            continue
        scalled_raw_img = cv2.resize(img,(32,32))
        img_har = w2d(img, 'db1',5)
        scalled_img_har = cv2.resize(img_har,(32,32))
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        X.append(combined_img)
        y.append(class_dict[celebirity_name])
```

## Step 3: Taining the Model

```python
# Import library
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0 )
```

#### Using SVM Model

```python
pipe = Pipeline([('scalar', StandardScaler()),('svm',SVC(kernel='rbf',C=10))])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
predict = pipe.predict(X_test)
```

```python
print(classification_report(y_test,predict))
```

### Using Grid Search to find the optimal model for this dataset

```python
#import library
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
#create dictionary to supply parameter for GridSearch
model_params ={
    'svm' : {
        'model' : SVC(gamma='auto', probability = True),
        'params':{
            'svc__C':[1,10,100,1000],
            'svc__kernel':['rbf', 'linear']
        }
    },
    'ramdom_forest':{
        'model' : RandomForestClassifier(),
        'params':{
            'randomforestclassifier__n_estimators':[1,5,10]
        }
    },
    'logistic_regression':{
        'model' : LogisticRegression(solver='liblinear', multi_class = 'auto'),
        'params' : {
            'logisticregression__C': [1,5,10]
        }
        
    }
    
}
```

```python
#create GridSearch
score = []
best_estimators = {}
import pandas as pd
for algo , mp in model_params.items():
    pipe = make_pipeline(StandardScaler(),mp['model'])
    clf = GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    score.append({
        'model': algo,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
    
df = pd.DataFrame(score, columns = ['model','best_score','best_params'])
df
```
